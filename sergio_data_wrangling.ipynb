```python
import pandas as pd
```

```python
df = pd.read_csv('./data/fred_data.csv')
```

```python
df.head()
```

```python
# Clean data
df_rename = df.rename({'DATE':'quarter'}, axis=1)
df_rename.head()
```

```python
# Update values in 'quarter' column to have the format: YYYYQ1-4. Example: 1970Q1

def map_quarter(s):
    quarter = ''
    month = s.split('-')[1]
    year = s.split('-')[0]
    
    if month == '01':
        quarter = '1'
    elif month == '04':
        quarter = '2'
    elif month == '07':
        quarter = '3'
    elif month == '10':
        quarter = '4'
        
    x = year + 'Q' + month
    
    return x

df_rename['quarter'] = df_rename['quarter'].apply(map_quarter)
df_rename.head()
```
